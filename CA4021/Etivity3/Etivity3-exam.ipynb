```python
import random 
import matplotlib.pyplot as plt

def throw_dice() :
        return random.randint(1, 4)
```

```python
rolls = []
for _ in range(100):
    rolled = throw_dice()                                          
    rolls.append(rolled)
    
print(rolls)
plt.hist(rolls, bins=range(1, 7), edgecolor='black', density=True)
plt.show()

dice_mean = sum(rolls) / 100
print(dice_mean)
            
```
